```python
#Data Type Conversion and Snapshot Date

from datetime import timedelta
import pandas as pd

# Convert ORDERDATE to datetime objects
df['ORDERDATE'] = pd.to_datetime(df['ORDERDATE'])

# Define a "today" date for Recency calculation (one day after the last order date)
snapshot_date = df['ORDERDATE'].max() + timedelta(days=1)
print(f"Analysis Snapshot Date: {snapshot_date}")
```

```python
#Group and Aggregate for RFM

# Calculate Recency, Frequency, and Monetary
rfm_table = df.groupby('CUSTOMERNAME').agg(
    # Recency: Find the max date, then subtract from the snapshot date and extract days
    Recency=('ORDERDATE', lambda x: (snapshot_date - x.max()).days),

    # Frequency: Count the number of unique order numbers (transactions)
    Frequency=('ORDERNUMBER', 'nunique'),

    # Monetary: Sum the total sales amount
    Monetary=('SALES', 'sum')
)

print("RFM Table Head:")
print(rfm_table.head())
```

```python
#Calculate RFM Scores (1-5)


# Recency: Smallest days gets the highest score (5), so we reverse the scoring list [5, 4, 3, 2, 1]
rfm_table['R_Score'] = pd.qcut(rfm_table['Recency'], q=5, labels=[5, 4, 3, 2, 1])

# Frequency: Highest frequency gets the highest score (5)
rfm_table['F_Score'] = pd.qcut(rfm_table['Frequency'].rank(method='first'), q=5, labels=[1, 2, 3, 4, 5])

# Monetary: Highest monetary value gets the highest score (5)
rfm_table['M_Score'] = pd.qcut(rfm_table['Monetary'], q=5, labels=[1, 2, 3, 4, 5])

# Convert scores to integers
rfm_table[['R_Score', 'F_Score', 'M_Score']] = rfm_table[['R_Score', 'F_Score', 'M_Score']].astype(int)

print("\nRFM Table with Scores:")
print(rfm_table[['R_Score', 'F_Score', 'M_Score']].head())
```

```python
#ðŸ·ï¸ Step 3: Assign RFM Segments
rfm_table['RFM_Score'] = rfm_table['R_Score'].astype(str) + \
                         rfm_table['F_Score'].astype(str) + \
                         rfm_table['M_Score'].astype(str)
```

```python
#Define and Map Segments

def assign_segment(df):
    if df['R_Score'] >= 4 and df['F_Score'] >= 4 and df['M_Score'] >= 4:
        return 'Champions' # Best customers: bought recently, buy often, spend the most
    elif df['R_Score'] >= 4 and df['F_Score'] >= 4:
        return 'Loyal Customers' # High Recency and Frequency
    elif df['R_Score'] >= 3 and df['F_Score'] >= 3 and df['M_Score'] >= 3:
        return 'Potential Loyalist' # Average scores, good potential
    elif df['R_Score'] >= 2 and df['F_Score'] <= 2:
        return 'At Risk' # Not recently bought, infrequent, but spent well previously
    elif df['R_Score'] == 1:
        return 'Lost Customers' # Worst scores: lowest Recency
    else:
        return 'Others'

rfm_table['Customer_Segment'] = rfm_table.apply(assign_segment, axis=1)

print("\nRFM Segments:")
print(rfm_table['Customer_Segment'].value_counts())
```

```python
#Preparation for CLV (Customer Lifetime Value)
# Calculate CLV pre-requisite metrics
clv_metrics = df.groupby('CUSTOMERNAME').agg(
    # T (Tenure): Age of customer in days (from first order to snapshot date)
    T=('ORDERDATE', lambda x: (snapshot_date - x.min()).days),

    # Monetary_Value: Average sales value per transaction (only for customers with >1 order)
    Monetary_Value=('SALES', 'mean')
)

# Merge back into the RFM table
rfm_table = rfm_table.join(clv_metrics)

# Final step: You would now use these metrics (Recency, Frequency, Monetary, T, Monetary_Value)
# as input for a CLV prediction library (e.g., `lifetimes` in Python).

print("\nCLV Ready Metrics:")
print(rfm_table[['Recency', 'Frequency', 'T', 'Monetary_Value']].head())
```

```python
#Advanced Aggregations
# Group by COUNTRY and summarize Sales
country_sales_summary = df.groupby('COUNTRY').agg(
    Total_Sales=('SALES', 'sum'),
    Num_Orders=('ORDERNUMBER', 'nunique')
).sort_values(by='Total_Sales', ascending=False)

print("\nTop 5 Countries by Sales:")
print(country_sales_summary.head())
```

```python
#Product Line Sales by Year (Pivot Table)

# Create a pivot table: PRODUCTLINE as rows, YEAR_ID as columns, SALES as values, using sum
product_year_pivot = pd.pivot_table(
    df,
    values='SALES',
    index='PRODUCTLINE',
    columns='YEAR_ID',
    aggfunc='sum',
    fill_value=0 # Fill missing (no sales) with 0
)

print("\nProduct Line Sales by Year:")
print(product_year_pivot)
```

```python
#Advanced Aggregations and Pivot Tables

# Group by Territory and Product Line, then aggregate total sales and average MSRP
territory_product_summary = df.groupby(['TERRITORY', 'PRODUCTLINE']).agg(
    Total_Sales=('SALES', 'sum'),
    Average_MSRP=('MSRP', 'mean'),
    Total_Quantity=('QUANTITYORDERED', 'sum')
).sort_values(by='Total_Sales', ascending=False)

print("\nTerritory-Product Sales Summary (Top 5):")
print(territory_product_summary.head())
```
