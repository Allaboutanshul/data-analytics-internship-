### Time Series Analysis
Since 'ORDERDATE', 'MONTH_ID', 'YEAR_ID', and 'QTR_ID' are present, the dataset allows the analysis of sales trends over time, seasonality, and growth patterns by month, quarter, or year.[1]

```python


# Assuming df is your DataFrame loaded with data

# Group sales by year, quarter, and month
sales_year = df.groupby('YEAR_ID')['SALES'].sum()
sales_quarter = df.groupby(['YEAR_ID', 'QTR_ID'])['SALES'].sum().unstack()
sales_month = df.groupby(['YEAR_ID', 'MONTH_ID'])['SALES'].sum().unstack()

# Plot total sales per year (bar chart)
sales_year.plot(kind='bar', figsize=(10,6), title='Total Sales by Year')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.show()

# Plot sales by quarter for each year (line plot)
sales_quarter.plot(kind='line', marker='o', figsize=(10,6), title='Sales by Quarter per Year')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.legend(title='Quarter')
plt.show()

# Plot sales by month for each year (line plot)
sales_month.plot(kind='line', marker='o', figsize=(12,7), title='Sales by Month per Year')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.legend(title='Month')
plt.show()

```

CREATE TIME SERIAL PLOT TO SHOW TRENDS

```python


# Convert ORDERDATE to datetime if not already done
df['ORDERDATE'] = pd.to_datetime(df['ORDERDATE'])

# Set ORDERDATE as the index for time series operations
df.set_index('ORDERDATE', inplace=True)

# Resample the sales data by day (you can use 'M' for monthly or 'Q' for quarterly)
daily_sales = df['SALES'].resample('D').sum()

# Plot the time series
plt.figure(figsize=(12,6))
plt.plot(daily_sales.index, daily_sales.values, label='Daily Sales', color='blue')
plt.title('Daily Sales Time Series')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```
